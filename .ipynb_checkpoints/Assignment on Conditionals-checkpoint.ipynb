##### Computing One Alphabet>>>

```python
letter=input("Enter an alphabet:")
letter_input=letter.lower().isalpha()
vowel="aeiou"
if letter in vowel and len(letter)==1:
    print(f'The letter "{letter}" is a vowel.')
elif letter=="Y" and len(letter)==1:
    letter_input=letter.lower().isalpha()
    print(f'The Letter "{letter}" is sometimes a vowel and sometimes a consonant.')
elif letter not in vowel and letter.isalpha():
    letter_input==letter.lower()
    print(f'The letter "{letter}" is a consonant.')
else:
    print(f'Invalid Entry!\nEntry must contain one character and must be a letter.')
```

##### Computing Numbers>>>

```python
number=int(input("Enter a number:"))
if number<0:
    print(f'The number "{number}" is negative')
elif number==0:
    print(f'The number is {number}')
else:
    print(f'The number "{number}" is positive') 
```

##### Computing Leap Year>>>

```python
year=int(input("Enter a year:"))
if year%4==0 and (year%100!=0 or year%400==0):
    print(f'{year} is a leap year')
else:
    print(f'{year} is not a leap year')
```

##### Comparing Two Numbers>>>

```python
number1=int(input("Enter a digit:"))
number2=int(input("Enter a digit:"))
if number1>number2:
    print(f'{number1} is greater than {number2}')
elif number1<number2:
    print(f'{number1} is less than {number2}')
else:
    print(f'{number1} is equal to {number2}')
```

##### Simple Calculator>>>

```python
number1=float(input("Enter a digit:"))
operator=input("Enter the operator (+, -, *, /):")
number2=float(input("Enter a digit:"))
if operator=="+":
    result=number1+number2
    print(f'{number1} + {number2}= {result}')
elif operator=="-":
    result = number1-number2
    print(f'{number1} - {number2} = {result}')
elif operator=="*":
    result= number1*number2
    print(f'{number1} * {number2} = {result}')
elif operator=="/":
    result=number1/number2
    print(f'{number1} / {number2} = {result}')
else:
    print(f'Error!\nInvalid Operator, Try Again.')
```

##### Age Categorization>>>

```python
age=int(input("Enter individual's age:"))
if age<0:
    print(f'Error! Invalid Entry, Try again')
elif age<=12:
    print(f'Individual is a child')
elif age<=19:
    print(f'Individual is a teenager.')
elif age<=64:
    print(f'Individual is an adult.')
else:
    print(f'Individual is a senior.')
```

```python
side1=int(input("Enter the length of the first side:"))
side2=int(input("Enter the length of the second side:"))
side3=int(input("Enter the length of the last side:"))
if side1==side2==side3:
    print(f'The triangle is equilateral.')
elif side1==side2 or side1==side3 or side2==side3:
    print(f'The triangle is an isosceles.')
else:
    print(f'The triangle is a scalene.')
```

##### Checking if password meets criteria>>>

```python
while True:
    password=(input("Enter the password:"))
    if len(password)<8:
        print("Error!: Password must be at least 8 characters.")
    elif password.lower()=="password" or password=="123456":
        print("Error! Password is too common. Please choose a sronger password.")
    else:
        print("Password is valid.")
        break
```

##### Checking characters in a string>>>

```python
string=input("Enter a string:")
string=string.replace(" ", "")
if string.isdigit():
    print("The string contains only digits.")
elif string.isalpha():
    print("The string contains only letters.")
else:
    print("The string contains a mix of both characters.")
```

##### Sentence punctuation>>>

##### Palindrome word>>>

```python
word=input("Enter a word:")
word_input=word.strip().lower()
if " " in word_input:
    print("Invalid entry! Enter a word.")
elif word_input==word_input[::-1]:
    print(f'"{word}" is a palindrome.')
else:
    print(f'"{word}" is not a palindrome.')
```

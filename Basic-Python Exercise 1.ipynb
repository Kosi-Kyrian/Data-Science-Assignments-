###### CALCULATES YEARS, WEEKS AND DAYS>>>

```python
total_days=input("enter number of days:")
years=int(total_days)//365
days_left=int(total_days)-(years*365)
weeks=days_left//7
days=days_left-(weeks*7)
print(f"{years}yr(s) {weeks}wk(s) {days}day(s)")
```

###### AREA AND CIRCUMFERENCE OF A CIRCLE>>>

```python
pi=3.14159
radius=input("enter radius:")
area=pi*int(radius)**2
circumference=2*pi*int(radius)
print(f"area of the circle=  {area}, circumference of the circle= {circumference}")
```

###### SUM, DIFFERENCE, PRODUCT, QUOTIENT, REMAINDER, POWER>>>

```python
x=input("enter first digit:")
y=input("enter second digit:")
sum=int(x)+int(y)
difference=int(x)-int(y)
product=int(x)*int(y)
quotient=int(x)//int(y)
remainder=int(x)%int(y)
power=int(x)**int(y)
print("sum=", sum, "difference=", difference, "product=", product, "quotient=", quotient, "remainder=", remainder, "power=", power)
```

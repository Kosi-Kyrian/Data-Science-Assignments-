##### 1a) Converting to Student Scores to Numpy array>>

```python
import numpy as np
```

```python
student_scores=[
    [85,92,78,96,88],
    [79,85,91,87,83],
    [92,88,95,91,89],
    [76,82,79,85,81],
    [88,91,87,93,90]
]
scores=np.array(student_scores)
print("Student Scores=",scores)

print("Array Shape=",scores.shape)

print("Array Dimension=",scores.ndim)
```

#### 2) Analyzing student performance across multiple subjects:

```python
subject_names=["Maths","Science","English","History","Art","PE"]
num_students,num_subjects=50,6
np.random.seed(456)
scores=np.random.randint(60,100,(num_students,num_subjects))
print(scores)
```

##### a) Studend who exceled in STEM (Maths + Science average > 90)

```python
stem_avg=scores[:,:2].mean(axis=1)
excel=np.where(stem_avg>90)[0]
print(f'Average scores of students in Maths and Science:{stem_avg}')
print("Students who exceled in Maths and Science (with average above 90)=",excel)
```

##### b) Identifying struggling students and subjects they struggle with (sunbjects < 70)>>

```python
for i in range(num_students):
    if np.any(scores[i]<70):
        weak_subjects=np.array(subject_names)[scores[i]<70]
        print(f"Student {i+1} struggles with {weak_subjects}")
```

##### c) Most challenging subject(lowest average) and best performance in that subject>>

```python
subject_avg=scores.mean(axis=0)
lowest_avg_index=np.argmin(subject_avg)
lowest_avg=subject_avg[lowest_avg_index]
most_challenging_sub=np.array(subject_names)[lowest_avg_index]
best_performance=scores[:,lowest_avg_index].max()
best_students=np.where(scores[:,lowest_avg_index]==best_performance)[0]
print("Subjects Average=",subject_avg)
print(f"Most challenging subject is '{most_challenging_sub}' with an average of '{lowest_avg}'")
print(f"Students {best_students} have the best performance in '{most_challenging_sub}' with score of '{best_performance}'")
```

#### 3) Analyzing a stock portfolio using Numpy's mathematical functions>>

```python
days=252
stocks=5
stock_names=['AAPL','GOOGL','MSFT','TSLA','AMZN']
initial_price=np.array([150,2800,300,800,3200])
np.random.seed(789)
daily_returns=np.random.normal(0.001,0.02,(days,stocks))
```

##### a) Basic Analysis>>

##### i) Cumulative Returns for each stock>>>

```python
daily_returns
```

```python
stock_cumulative_returns=np.cumprod(1+daily_returns,axis=0)-1
final_stock_cum_returns=stock_cumulative_returns[-1]
print("Cumulative Returns=")
for i in range(stocks):
    print(stock_names[i],"===", round(final_stock_cum_returns[i],3))
```

##### ii) Annualized volatility (standard deviation * sqrt(252))>>>

```python
volatility=np.std(daily_returns,axis=0)*np.sqrt(252)
print("Annualized volatility:")
for i in range(stocks):
    print(stock_names[i],">>>", round(volatility[i],3))
```

##### iii) Best and worst performing stock>>>

```python
best_stock_index=np.argmax(final_stock_cum_returns)
best_stock=(final_stock_cum_returns)[best_stock_index]
worst_stock_index=np.argmin(final_stock_cum_returns)
worst_stock=(final_stock_cum_returns)[worst_stock_index]
print("Best performing stock:")
print(np.array(stock_names)[best_stock_index], "with", round(best_stock,3), "Stocks")
print("Worst performing stock:")
print(np.array(stock_names)[worst_stock_index], "with", round(worst_stock,3),"Stocks")
```

##### b) Risk Metrics>>>

##### i) Calculating sharpe ratio (return/volatility) for each stock>>>

```python
sharpe_ratio=final_stock_cum_returns/volatility
print ("Sharpe Ratio:")
for i in range(stocks):
    print(stock_names[i], ">>>", round(sharpe_ratio[i],3))
```

##### ii) Maximum drawdown (largest_peak_to_trough_decline)>>>

```python
prices=initial_price*(1+stock_cumulative_returns)
max_drawdown=[]
for i in range(stocks):
    peak=prices[:,i].max()
    trough=prices[:,i].min()
    drawdown=(trough-peak)/peak
    max_drawdown.append(drawdown)
print("Maximum Drawdown:")
for i in range (stocks):
    print(stock_names[i], ">>>", round(max_drawdown[i],3))
```

##### iii) Correlation matrix between stocks>>>

```python
correlaion_matrix=np.corrcoef(daily_returns.T)
print("Correlation Matrix:")
print(correlaion_matrix)
```

```python
def create_email(first_name,last_name,domain="gmail.com"):
    email=f"{first_name.lower()}{last_name.lower()}@{domain.lower()}"
    return email
```

9. Create a funtion "count_vowels" that takes a string and returns the number of vowels in it.

```python
def count_vowels(string):
    vowels="aeiou"
    count=0
    for letter in string.lower():
        if letter in vowels:
            count+=1
            print(letter)
    return count
```

```python
count_vowels("George")
```

11. Write a function "grade_calculator" that takes a score (0-100) and returns the letter grade (A, B, C, D, E, F).

```python
def grade_calculator(score):
    if score<0 or score>100:
        return "Invalid entry!"
    elif score>=70:
        return "A"
    elif score>=60:
        return "B"
    elif score>=50:
        return "C"
    elif score>=45:
        return "D"
    elif score>=40:
        return "E"
    else:
        return "F"
```

```python
grade_calculator(15)
```

12. Write a function "file_extension" that extracts the file extension from a file name, handling cases where there might be no extension.

```python
def file_extension(file_name):
    parts=file_name.split('.')
    if len(parts)>1:
        return f'.{parts[-1]}'
    else:
        return "File has no extension."
```

```python
file_extension("Kosi.txt")
```

```python
file_extension("Kosi.ipynb")
```

```python
file_extension("Kosi")
```

13.Write a function "safe_get_item" that tries to access a list item by index and returns None if the index is out of bounds.

```python
def safe_get_item(list,index):
    try:
        return list[index]
    except IndexError:
        return "None"
```

```python
courses=["data science", "data analysis", "cyber security", "web development"]
```

```python
safe_get_item(courses,2)
```

```python
safe_get_item(courses,4)
```

```python
def safe_get_items(list,index):
    if index in range(len(list)):
        return list[index]
    else:
        return "None"
```

```python
safe_get_items(courses,0)
```

```python
safe_get_items(courses,5)
```

14. Create a function "clean_data" that takes a list of mixed data types and returns only the numeric values as a new list.

```python
def clean_data(mixed_data_types):
    numerics=[]
    for datatypes in mixed_data_types:
        if isinstance(datatypes, (int, float)) and not isinstance(datatypes, bool):
            numerics.append(datatypes)
    return numerics
    
```

```python
mixed_data=[1, 'hello', 3.5, True, None,'5', 10]
clean_data(mixed_data)
```

15. Write a function "analyze_grades" that takes a dictionary of students names and their scores lists, returning satistics(average, highest, lowest) for each student.

```python
def analyze_grades(grades):
    statistics={}
    for students, scores in grades.items():
        statistics[students]={
            "average": round(sum(scores)/len(scores),2),
            "highest": max(scores),
            "lowest": min(scores)}
    return statistics
```

```python
grades={"John":[85, 90, 78],"Alice": [92,88,95], "Bob": [75,80,70]}
```

```python
analyze_grades(grades)
```
